```python
import os


def list_files(path):
    for f in os.listdir(path):
        pathname=os.path.join(path, f)
        print(pathname)
        if os.path.isfile(f):
            print(pathname)
        if os.path.isdir(f):
            list_files(f)
path="c:\\users"
list_files(path)
```
