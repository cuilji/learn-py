```python
import requests
from lxml import etree
import time

url='https://www.guancha.cn/'
r=requests.get(url)
r.encoding='utf-8'
html=r.text
selector=etree.HTML(html)
#title=selector.xpath('//title/text()')
links=selector.xpath("//h4[@class='module-title']/a/@href")
#print(len(title))
domain='http://www.guancha.cn'
for l in links:
    #print(domain + l)
    res=requests.get(domain + l)
    print(res.ok)
    time.sleep(2)
```
