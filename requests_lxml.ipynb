```python
import requests
from lxml import etree

url='https://www.guancha.cn/'
r=requests.get(url)
r.encoding='utf-8'
html=r.text
selector=etree.HTML(html)
#title=selector.xpath('//title/text()')
links=selector.xpath("//h4[@class='module-title']/a/@href")
#print(len(title))
for l in links:
    print(l)
    print('\n')
```
